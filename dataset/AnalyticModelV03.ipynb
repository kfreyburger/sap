```python
import requests
import pandas as pd
import numpy as np
import random
from datetime import datetime
from datetime import date
```

```python
saison = pd.read_excel("https://raw.githubusercontent.com/kfreyburger/sap/main/dataset/AnalyticModelV03.xlsx", sheet_name = "Saison",header = 0, engine='openpyxl')
```

```python
anzahl = 10

erstesJahr = 2020
letzesJahr = 2020

# array zum Testen der Monatsverteilung
v = np.zeros(13)
start = datetime.now()


# Werte werden in Listen erstellt und unten zu einem Dataframe zusammengeklebt
dates = []
salesquan=[]

for index in range(0,anzahl):
  jahr = 2020 + int(random.random()*(letzesJahr - erstesJahr + 1))
# Doku: Der Boolsche Ausdruck lÃ¤sst nur die EintrÃ¤ge Ã¼brig, welche die Bedingung erfÃ¼llen
# [-1:] nimmt davon den letzen Eintrag; und ['Monat'] davon wiederum nur den Monat
  monat = int(saison[saison['Normiert']<random.random()][-1:]['Monat'])
  v[monat] = v[monat]+1
  if monat == 12:
    lastday = 31
  else:  
    date1 = date(jahr,monat,1)
    date2 = date(jahr,monat+1,1)
    lastday = (date2-date1).days
  tag = 1+ int(random.random()*lastday)
  dates.append(date(jahr, monat, tag).strftime("%Y-%m-%d"))
  
  salesquan.append(int(1+10*random.random()))

# man sieht, dass die Gewichte rauskommen!!!
print(v/anzahl*38.5)
```

```python
sales = pd.DataFrame({'Date' : dates, 'Salesquan': salesquan})
print(sales)
```

```python
anz_orders = 5
orders=[]
items = []
dates = []
salesquan=[]
uom=[]
customers = []
products=[]
revenue=[]
curr=[]
discount=[]
cost=[]

for order in range(0,anz_orders):
#   Kunden aus der 2. Hälfte genommen --> Deutsche    
    customer = df_customer[df_customer['Normiert']<0.5+random.random()/2][-1:]
    tag = 1+ int(random.random()*31)
    salesdate = pd.to_datetime(date(2020, 1, tag))
    anz_items = int(1+10*random.random())
    for item in range(0,anz_items):
        orders.append(order)
        items.append((item+1)*10)
        customers.append((int)(customer['Customer'].values[0]))
        dates.append(salesdate)
        product = df_product[df_product['Normiert']<random.random()][-1:]
        products.append(product['Product'].values[0])
        quan = int(1+10*random.random())
        uom.append('ST')
        salesquan.append(quan)
#    Faktor EUR aus 2019 genommen 1.16964551
        rev = round(product['Salesprice'].values[0]*quan*1.16964551,2)
        revenue.append(rev)
        if (customer['Customer'].values[0]=='US'):
            curr.append('USD')
        else:
            curr.append('EUR')
        discount.append(round(rev*customer['Rabatt'].values[0],2))
#    Faktor CostKoef aus 2019    1.097636        
        cost.append(round(product['InternalPrice'].values[0]*quan*1.097636,2))
        
df_sales = pd.DataFrame({'OrderNumber':orders,'OrderItem': items,'SalesDate' : dates, 'Customer':customers,'Product':products,'SalesQuantity': salesquan, 'UnitOfMeasure': uom,'Revenue':revenue,'Currency':curr ,'Discount':discount,'CostsUSD':cost})
df_sales
```

weiter OrderNumber 

```python
df_sales = df_sales.sort_values(by=['SalesDate','OrderNumber','OrderItem'])
df_sales
```

```python
first_number = 120001
# durch ignore_index = True wird das Ergebnis neu durchnummeriert, iloc macht alle Zeilen und die erste Spalte
df_numbers = df_sales.drop_duplicates(subset='OrderNumber',ignore_index=True).iloc[:,[0]]
# eine neue Spalte macht man am besten mit Assign
df_numbers = df_numbers.assign(NewOrderNumber=df_numbers.index+first_number)
df_numbers
```

```python
# Versuch in weniger Schritten ging schief, warum auch immer
df_sales = df_sales.merge(df_numbers,on='OrderNumber')
df_sales['OrderNumber']=df_sales['NewOrderNumber']
df_sales = df_sales.drop(columns=['NewOrderNumber'])
df_sales
```

```python
# deutsches Datum
df_sales['SalesDate'] = df_sales['SalesDate'].dt.strftime("%d.%m.%Y")
```
